# Simple math on DataFrame

```python
df.count() 
#row NA/null values
# 计算DataFrame中每列非缺失（非NaN）元素的数量
```

```python
df.cummax() #(cols default axis)
#用于计算DataFrame中每列的累积 "最大值 "
#返回一个新 DataFrame
#其中的每个元素表示在当前行之前（包括当前行）的所有行中每列的最大，值大則覆蓋之
#因此：最後一列顯示的為每項目中的最大值
```

```python
df.cummin() #(cols default axis)
#用于计算DataFrame中每列的累积 "最小值 " 
#返回一个新 DataFrame
#其中的每个元素表示在当前行之前（包括当前行）的所有行中每列的最小，值小則覆蓋之
#因此：最後一列顯示的為每項目中的最小值
```

```python
df.cumsum() #(cols default axis)
#计算DataFrame中每列的累积总和
#返回一个新的DataFrame
#其中的每个元素表示在当前行之前（包括当前行）的所有行中每列的 累积 "总和 "
```
