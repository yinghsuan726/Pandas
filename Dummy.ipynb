### 1.`.str.replace()` : 先拿掉 `'$',','`等特殊符號

# Create **Dummy Variables** in Pandas : `.get_dummies()`

> **將 categorical data 轉換成 dummy variables，方便運算**

- **注意 `pd.get_dummies` “不會” 保留原項目（因為是“轉換”）**
  - **欲保留原項目 : 則個別轉換另存後，與原表格進行合併（ `pd.concat` ）**

**`pandas.get_dummies(data, columns=None, prefix=None, prefix_sep='_', drop_first=False, dtype=None)
`**
| param        | meaning                                                 | =                                                                 |
| ------------ | ------------------------------------------------------- | ----------------------------------------------------------------- |
| `data`       | 要進行轉換的 DF/ Series                                 |                                                                   |
| `columns`    | 用于指定要进行虚拟变量转换的列名                        | |
|`prefix`|前綴，區分原始列名|`None`: 不加前綴 (default) |
|`prefix_sep`|指定前綴與原列名間的分隔號|`_`： default|
| `drop_first` | 因只需要(k-1)個 dummies，為避免共線性刪除第一個 dummy？ | `False` : (default)                                                 |
|              |                                                         | `True` : 刪除每個分類變量的首個 dummy                               |
| `dtype`      | 指定數據類型？                                          | `None` : 自動推斷數據類型                                          |



### 1. `pd.get_dummies` : 獲取虛擬變數

```python
pd.get_dummies(Mar_1.Education)
## get [] levels of Dummies in the object 
# We can get 5 levels in Education
#* Mar_1. == Mar_1[]
```

```python
pd.get_dummies(Mar_1.Education, prefix='Education', drop_first=True)
# 函数生成的虚拟变量列显示为 "True" ,"False" 而不是 0 , 1
# 通常这是因为 DataFrame 的数据类型被解释为布尔型（Boolean）
```

### 2. Dummies = (k-1) variables can represent all info, delete 1 dummy
##### 
- 原本：`.iloc[:,1:]`
  - (k-1) : set `'2nd Cycle'` as reference group
  - .iloc[ALL rows(:), Cols starting in position 1 through the last one]`**
- fast : `.get_dummies(,drop_first=True)` : 刪除每個類別變數的第一個dummy


```python
pd.get_dummies(Mar_1.Education).iloc[:,1:]
## we just need (k-1) Dummies 
## way1 - iloc 
#* better : .get_dummies(,drop_first=True) 
```

```python
pd.get_dummies(Mar_1.Education, drop_first=True)
## way2 - drop_first
#* .iloc[:,1:]
```

### Set `prefix（前綴）` to easily tell each dummy


```python
pd.get_dummies(Mar_1.Education,prefix='Education').iloc[:,1:]
pd.get_dummies(Mar_1.Education, prefix='Education', drop_first=True)
## get prefix 
#//`(,prefix_sep='_')` 
```

```python
# 函数生成的虚拟变量列显示为 "True" ,"False" 而不是 0 , 1
# 通常这是因为 DataFrame 的数据类型被解释为布尔型（Boolean）
# 若想要虛擬變量以0, 1表示，+`.astype(int)`
Education_Dummies = pd.get_dummies(Mar_1.Education, prefix='Education', drop_first=True).astype('uint8')
Education_Dummies
```

## 資料結構合併：`pd.concat`

> **将多个 Pandas 数据结构（例如数据框或系列）沿指定的轴（通常是行轴或列轴）进行连接或合并**

### (p.s.) 是 “ Feature Engineering 特徵工程 ” 之一
> **try to make more cols of ur DTset to feature ur DT**

**`pd.concat(objs, axis=0, join='outer', ignore_index=False, keys=None, levels=None, names=None, verify_integrity=False, sort=False, copy=True)
`**
| param           | meaning                             | =                                  |
| -------------- | ----------------------------------- | ---------------------------------- |
| `objs`         | 欲進行合併的表格們                  | `'Name' `                            |
| `axis`         | 指定連接的軸（# `0`，行）       | # `0`：沿行合併，每行變長            |
|                |                                     | `1`：沿列合併，每列變長            |
| `join`         | 外連接？(# `'outer'`)         | # `outer`：外連接，保留所有行/列     |
|                |                                     | `inner`：內連接，只保留共有的行/列 |

較少用的部分：

| param               | meaning                            | =          |
| ---------------- | -------------------------------- | ------- |
| `ignore_index`     | 是否重新產生索引？(#  `False`)     | #  `False` |
| `keys`             | 是否為每個資料結構加標籤區分它們？ | #  `'None'`        |
| `levels`           | 指定層級索引名稱                   | # `None`   |
| `names`            | 指定索引級別的名稱                 | # `None`   |
| `verify_integrity` | 连接时检查结果是否有重复的索引?    | # `False`  |
| `sort`             | 是否對結果的索引進行排序？         | # `False`  |
| `copy`             | 在連接中複製資料？                 | # `True`   |

# Dealing w/ Dates & Times : `pd.to_datetime()`
> **用于将输入数据转换为日期时间对象的函数。**

`pd.to_datetime(arg, format=None, errors='raise', utc=False, infer_datetime_format=False, dayfirst=False, yearfirst=False, exact=True, unit=None, origin='unix', cache=False)`

| param                   | meaning                                      | =                                     |
| ----------------------- | -------------------------------------------- | ------------------------------------- |
| `arg`                   | "arguments"（参数），要转换为日期时间的数据  |                                       |
| `errors`                | 决定如何处理解析错误                         | # `'raise'` : 引發KeyError            |
|                         |                                              | `'ignore'`                            |
| `utc`                   | 返回的日期时间将使用 UTC 格式                | # `'False'`                           |
| `infer_datetime_format` | 尝试推断日期时间字符串的格式，以提高解析性能 | # `'True'`                            |
| `dayfirst`              | 日/月/年的格式（例如，28/01/2022）           | # `'False'`                           |
| `yearfirst`             | 年/月/日的格式（例如，2022/01/28）           | # `'False'`                           |
| `exact`                 | 要求输入的日期时间字符串与指定的格式完全匹配 | # `'True'`                            |
| `unit`                  | 指定输入数据的时间单位                       | # `'ns'` : 納秒                        |
| `origin`                | 解析数值数据的参考日期                       | # `'unix'` : 表从1970年1月1日开始计算 |
| `cache`                 |启用日期时间解析的缓存，以提高性能|# `'False'`|

##### `.loc[]` v.s. `.iloc[]`
|              | `.loc[]`                                             | `.iloc[]`                                      |
| ------------ | ---------------------------------------------------- | ---------------------------------------------- |
| 選擇資料方式 | `'標籤索引'`                                         | `'整數'`                                       |
| code         | `df.loc['行标签', '列标签']`                         | `df.iloc[行位置, 列位置]`                      |
| 切片行為     | 包括起始和结束标签在内的所有行和列都会被包括在结果中 | 起始位置包括在结果中，但结束位置不包括在结果中 |
| 用處 |不規則選取|規則選取某區塊|
